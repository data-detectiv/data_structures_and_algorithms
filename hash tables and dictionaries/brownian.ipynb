```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_brownian_motion(T=1, N=1000, seed=None):
    """
    Simulates a Brownian motion path.
    
    Parameters:
        T (float): Total time (default 1)
        N (int): Number of time steps (default 1000)
        seed (int): Random seed for reproducibility
    
    Returns:
        t (np.array): Time points
        W (np.array): Brownian motion path
    """
    if seed is not None:
        np.random.seed(seed)
    
    dt = T / N  # Time step
    t = np.linspace(0, T, N+1)  # Time grid
    dW = np.random.normal(0, np.sqrt(dt), N)  # Wiener increments
    W = np.cumsum(dW)  # Cumulative sum of increments
    W = np.insert(W, 0, 0)  # Start at 0
    
    return t, W

# Simulate and plot
t, W = simulate_brownian_motion(T=1, N=1000, seed=42)

plt.figure(figsize=(10, 6))
plt.plot(t, W, label="Brownian Motion Path")
plt.title("Simulated Brownian Motion (Wiener Process)")
plt.xlabel("Time (t)")
plt.ylabel("W(t)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
memo = []
def fib(n, memo):
    if memo[n] != None:
        return memo[n]
    if n <= 2:
        res = 1
    res = fib(n - 1) + fib(n - 2)
    memo[n] = res
    return res
```
